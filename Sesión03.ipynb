```python
def resta(a=None, b=None):
  if a == None or b == None:
    raise Exception("Error, no se puede realizar la resta sin las dos variables requeridas")
  return a - b

print(resta(2,5))

def suma(a=None, b=None):
  if a == None or b == None:
    raise Exception("Error, no se puede realizar la sumar sin las dos variables requeridas")
  return a + b

def multiplicacion(a=None, b=None):
  if a == None or b == None:
    raise Exception("Error, no se puede realizar la multiplicar sin las dos variables requeridas")
  return a*b

def operaciones(numero1, numero2):
  operacion_suma= suma(numero1,numero2)
  operacion_resta= resta(numero1,numero2)
  operacion_multiplicacion= multiplicacion(numero1,numero2)
  return operacion_suma , operacion_resta , operacion_multiplicacion

numero=5
numero2=10
resultado_operaciones = operaciones(numero,numero2)

print("La suma de ",numero, "y",numero2,"es:",resultado_operaciones[0])
print("La resta de ",numero, "y",numero2,"es:",resultado_operaciones[1])
print("La multipliacion de ",numero, "y",numero2,"es:",resultado_operaciones[2])
```
